1. Составить регулярные выражения и выделить из приведенного выше текста:
- все номера  телефонов;
- все адреса электронной почты;
-  все метки времени (часы минуты);
- все улицы/проспекты/бульвары и т.д.
- всех водителей с фамилией и инициалами.

```python
import re

s = """Водитель автомобиля Lada 2107 гос номер AH220H161 под
управлением водителя Сидоров И.С. 1927 г.р.,

9.09.2023 в 18:23
двигаясь по улице Сарьяна от бульвара Киселева в направлении ул. Ленина,
допустил столкновение с автомобилем Мерседес G600 с госномером OO777O61 под управлением водителя Староминская Н.К. 1998 г.р.
В результате столкновения второй автомобиль задел припаркованный автомобиль Лада Веста госномер ам323р23, принадлежащий водителю Суханов К.З.
Также в результате столкновения летящими обломками бампера поврежден а/м Фольксваген Жук гос. номер РО123C 123
под управлением водителя С.Н. Бублик Пострадавших в результате ДТП нет.

на месте происшествия работал старший инспектор ГИБДДД
лейтенант Заверюха Г.Е.
телефон 8(928)124-12-17
адрес. zaveruha@mvd.ru
протокол составлен 10/09/23 в 4:01
Водителям надлежит явиться в управление по адресу ул. Сидоренко 25 13-09-2023 в 9:00 ."""
# все номера  телефонов
pattern1_1 = r'(8\D?\d{3}\D?\d{3}\D?\d{2}\D?\d{2})'
# все адреса электронной почты
pattern1_2 = r'\w+@\w+\.ru'
# все метки времени (часы минуты);
pattern1_3 = r'\D\d+\:\d+'
# все улицы/проспекты/бульвары и т.д
pattern1_4 = r'((?:ул\.?|улиц\w?|бульвар\w?|проспект\w?)\s\w+)'
# всех водителей с фамилией и инициалами
pattern1_5 = r'((?:\w\.\w\.\s[А-Я]{1}[а-я]+)|(?:[А-Я]{1}[а-я]+\s\w\.\w\.))'

for item1 in re.findall(pattern1_1,s):
  print(item1)
print('\n')
for item2 in re.findall(pattern1_2,s):
  print(item2)
print('\n')
for item3 in re.findall(pattern1_3,s):
  print(item3)
print('\n')
for item4 in re.findall(pattern1_4,s):
  print(item4)
print('\n')
for item5 in re.findall(pattern1_5,s):
  print(item5)
```

2. Разработать на языке Python функцию, которая с применением регулярных выражений преобразует строку вида ‘N т. M кг.’ в число, соответствующее массе в килограммах

```python
# так же отрабатывает в случае отсутсвия кг или т
s2 = "Масса 4 т. 342 кг."
patternNum = r'\d+'
pattern2_1 = r'\d+ т\.'
pattern2_2 = r'\d+ кг\.'


for t in re.findall(pattern2_1,s2):
  print(t)
for kg in re.findall(pattern2_2,s2):
  print(kg)

if len(t) < 2:
  Lt = 0
else:
  t = re.findall(patternNum,t)
  Lt = int(t[0])
if len(kg)  < 2:
  Lkg = 0
else:
  kg = re.findall(patternNum,kg)
  Lkg = int(kg[0])

Mas = (Lt * 1000) + Lkg
print('\n')
print(Mas)

```
